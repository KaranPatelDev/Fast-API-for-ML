```python
import pickle
import joblib
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Read the data

```python
# df = pd.read_csv('./housing.csv')
# df
```

```python
df = pd.read_csv('./housing.csv').iloc[:,:-1].dropna()
df
```

```python
x = df.drop(columns=['median_house_value'])
y = df.median_house_value.copy()

print(x.shape, y.shape)
```

## Train the Model

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
r2_score(y, y_pred)
```

```python
model
```

## Pickle

```python
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
with open('model.pkl', 'rb') as f:
    picke_model = pickle.load(f)
```

```python
picke_model
```

```python
y_pred_pickle = picke_model.predict(x)
y_pred_pickle
```

```python
r2_score(y, y_pred_pickle)
```

## Joblib

```python
joblib.dump(model, 'model_joblib')
```

```python
joblib.load('model_joblib')
```

```python
y_pred_joblib = picke_model.predict(x)
y_pred_joblib
```
